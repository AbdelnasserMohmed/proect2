```python
plt.figure(figsize=(10, 7))
plt.boxplot(df['Scheduled'], vert=False)
plt.xlabel('x Scheduled ')
plt.ylabel('y Scheduled')
plt.title('Boxplot with Outliers')
plt.show()
```

```python
carrier_counts = df['carrier'].value_counts()


threshold = 25000
small_carriers = carrier_counts[carrier_counts < threshold]
other_carriers_count = small_carriers.sum()

# Replace carriers with counts less than 5 with "Other"
counts = carrier_counts[carrier_counts >= threshold]
counts['Other'] = other_carriers_count

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution ')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
